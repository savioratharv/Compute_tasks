```python
d1 = dataset[['artist']].groupby(['artist'])['artist'].count().nlargest(10) #finding top 10 artists by count
d1
```

```python
d2=d1.to_dict() 
l1=list(d2.keys()) #converting to a list and making a list of artists
print(l1)
```

```python
d3=dataset[['artist','tempo']]
l2=d3.values.tolist() #creating a list of all artists and the tempos of their different songs
print(l2)
```

```python
l4=[] #averages of artists will be appended here
import statistics
for i in l1: #looping through top 10 artists
  l3=[] #this list collects all the tempos of a specific artist (clears every loop)
  for j in l2: #looping through the previously created list of all the tempos
    if(i==j[0]): 
      l3.append(j[1]) #appending tempos of only the artists within the top 10
  l4.append(statistics.mean(l3)) #finding the mean of the tempos of a artist and appending it to l4
l5=zip(l1,l4) #zipping the list of top 10 artists with their average tempos
l6=dict(l5) #converting zip to dictionary
print(l6)
```

```python
new_d=pd.DataFrame.from_dict([l6]) #converting dictionary to dataframe
new_d
```
