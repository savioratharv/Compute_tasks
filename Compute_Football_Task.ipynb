```python
sns.scatterplot(x=df_pa['standing_tackle'],y=df_pa['marking'],palette='mako',hue=df_pa['overall_rating'])  #here we find that plauers with high marking and high standing tackle habe a high overall rating but for players with mid stancing tackle and marking, overall rating is not dependent
```

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.colors import ListedColormap

fig = plt.figure(figsize=(20,20))
ax = Axes3D(fig, auto_add_to_figure=False)
fig.add_axes(ax)
 
# get colormap from seaborn                                         #same inference as previous but graph is 3d for better visualization
cmap = ListedColormap(sns.color_palette("mako", 256).as_hex())      #here we find that plauers with high marking and high standing tackle habe a high overall rating but for players with mid standing tackle and marking, overall rating is not dependent

# plot
sc = ax.scatter(df_pa['standing_tackle'],df_pa['marking'], df_pa['overall_rating'], s=40, c=df_pa['standing_tackle'], marker='o', cmap=cmap, alpha=1)
ax.set_xlabel('standing_tackle')
ax.set_ylabel('marking')
ax.set_zlabel('overall_rating')
```

```python
fig = plt.figure(figsize=(20,20))
ax = Axes3D(fig, auto_add_to_figure=False)
fig.add_axes(ax)

# get colormap from seaborn
cmap = ListedColormap(sns.color_palette("mako", 256).as_hex())   #here we find that plauers with high marking and high sliding tackle habe a high overall rating but for players with mid sliding tackle and marking, overall rating is not dependent

# plot
sc = ax.scatter(df_pa['sliding_tackle'],df_pa['marking'], df_pa['overall_rating'], s=40, c=df_pa['sliding_tackle'], marker='o', cmap=cmap, alpha=1)
ax.set_xlabel('sliding_tackle')
ax.set_ylabel('marking')
ax.set_zlabel('overall_rating')
```
